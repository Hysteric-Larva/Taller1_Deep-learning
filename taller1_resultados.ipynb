```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
```

# Funciones a utilizar

```
def plot_corr(df):
    corr = df.corr()
    sns.set_style("darkgrid")
    plt.figure(figsize=(9,7))
    sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values,cmap="Spectral",annot = True)
    plt.show()

def plot_heat_rsq_test(df):
    pivot = df.pivot("Neuronas Totales", "Capas", "RSQ TEST")
    plt.figure(figsize=(9,7))
    sns.heatmap(pivot,cmap="Spectral",annot = True)
    plt.show()

def plot_heat_tiempo(df):
    pivot = df.pivot("Neuronas Totales", "Capas", "Tiempo")
    plt.figure(figsize=(9,7))
    sns.heatmap(pivot,cmap="Spectral",annot = True)
    plt.show()

def plot_heat_accuracy(df):
    pivot = df.pivot("Neuronas Totales", "Capas", "Accuracy")
    plt.figure(figsize=(9,7))
    sns.heatmap(pivot,cmap="Spectral",annot = True)
    plt.show()

def graficar(x,y):
    sns.set_style("darkgrid")
    plt.figure(figsize=(8,6))
    plt.plot(x,y,'.',c='darkblue')
    plt.axis('auto')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

def grafica_blobs(X,y):
    sns.set_style("darkgrid")
    plt.figure(figsize=(8,6))
    plt.scatter(X[:,0],X[:,1],c=y,cmap='cividis')
    plt.show()
```

# Objetivo del problema

Consideremos una arquitectura base $A$ con $n$ neuronas y $L$ capas, y supongamos que se redistribuyen las neuronas aumentando la cantidad de capas.

Buscamos responder a la pregunta: ¿Será posible aprender mejor y más rápido la tarea?.

# Caracteristícas de los modelos estudiados

## Modelo

<ol>
  <li>Red secuencial</li>
  <li>Capas densas</li>
  <li>Activaciones tipo Relu</li>

</ol>


## Entrenamiento

<ol>
  <li>Loss en regresión: error cuadrático medio (mse)</li>
  <li>Loss en clasificación: categorical cross entropy</li>
  <li>Optimizador: RMSprop </li>
  <li>Learning rate = 0.001 </li>
  <li>Early stopping: monitor = 'val_loss', patience = 10</li>
  <li> 1000 épocas de entrenamiento máximo </li>
  <li> Batch size = 32</li>
</ol>

## Primera variación de la arquitectura del modelo

<ol>
  <li>Considerar un red con una capa oculta con $2^p$ neuronas</li>
  <li>Considerar una nueva red con dos capas ocultas con $2^{p-1}$ neuronas cada una</li>
<li>Repetir este proceso hasta obtener una red que tiene todas las capas con $2$ neuronas</li>
</ol>


# Regresión simple

Inicialmente consideraremos un data set sintético. Para ello generaremos 1000 puntos en el plano, cuyo comportamiento sea parabólico.

```
x = np.linspace(-10,10,1000) 
y = [(i**2)+6*np.random.randn() for i in x] 
graficar(x,y)
```
